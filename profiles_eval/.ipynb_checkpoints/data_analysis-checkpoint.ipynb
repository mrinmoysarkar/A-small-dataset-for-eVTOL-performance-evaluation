```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
%matplotlib inline
```

```python
all_file_paths = []
for (root,dirs,files) in os.walk('.', topdown=True):
    for f in files:
        if 'profile' in f:
            path = os.path.join(root,f)
            all_file_paths.append(path)
```

# Mission profile fesible and infesible

```python
fesible_profile = None
fesible_speed = None
fesible_t = None

infesible_profile = None
infesible_speed = None
infesible_t = None
infesible_segment_t = None

infesible_segment = None
infesible_labels = None

max_range_fes = 0
max_range_infes = 0

for file in all_file_paths:
    df = pd.read_csv(file)
    t = df['time']
    range_profile = df['range']
    current_max_range = max(range_profile)
    if current_max_range>30:
        if 'True' in file and current_max_range>max_range_fes:
            max_range_fes = current_max_range
            fesible_profile = df['altitude']
            fesible_speed = df['speed']
            fesible_t = t
            
        elif 'False' in file and current_max_range>max_range_infes:
            max_range_infes = current_max_range
            infesible_profile = df['altitude']
            infesible_speed = df['speed']
            infesible_t = t
            infesible_segment = df['converged']
            infesible_labels = df[~df['converged']]['label']
            infesible_segment_t = df[~df['converged']]['time']
            
plt.figure(figsize=(5,3),dpi=200)            
plt.plot(fesible_t, fesible_profile, color='blue',linewidth=2,label="feasible mission profile", 
         marker='o', markersize=10)
plt.plot(infesible_t, infesible_profile, color='red',linewidth=2,label="infeasible mission profile",
         marker='*',linestyle='dashed')
plt.xlabel('Time (min)',fontsize=12)
plt.ylabel('Altitude (ft)',fontsize=12)
plt.legend(fontsize=10)
start = infesible_segment_t.iloc[0]
end = -1
for i in range(len(infesible_segment_t)-1):
    if infesible_segment_t.index[i+1] - infesible_segment_t.index[i] != 1:
        plt.axvspan(start, infesible_segment_t.iloc[i], facecolor='#ff00004f')
        start = infesible_segment_t.iloc[i+1]
plt.axvspan(start, infesible_segment_t.iloc[-1], facecolor='#ff00004f') 
plt.grid()
```

# Energy analysis

```python
fesible_energy = []
fesible_mission_range = []

for file in all_file_paths:
    if 'True' in file:
        df = pd.read_csv(file)
        fesible_energy.append(max(df['energy'])-min(df['energy']))
        fesible_mission_range.append(max(df['range'])-min(df['range']))
            
fesible_energy = np.array(fesible_energy)/3600

plt.figure(figsize=(5,3),dpi=200) 
plt.stem(fesible_mission_range,fesible_energy)
plt.xlabel("Range (mi)")
plt.ylabel("Consumed Battery Energy (W-hr)")
```

# Voltage analysis

```python
fesible_voltage = []
fesible_mission_range = []

for file in all_file_paths:
    if 'True' in file:
        df = pd.read_csv(file)
        fesible_voltage.append(min(df['voltage']))
        fesible_mission_range.append(max(df['range'])-min(df['range']))
            


plt.figure(figsize=(5,3),dpi=200) 
plt.stem(fesible_mission_range,fesible_voltage)
plt.xlabel("Range (mi)")
plt.ylabel("Battery Voltage (volt) \n at the end of the mission")
```

# C-Rating analysis

```python
labels = ['hover_climb', 'dep_transition', 'departure_terminal_procedure', 'accel_climb', 'cruise',
          'decel_descend', 'arrival_terminal_procedure', 'app_transition','hover_descent']

c_rating_in_segment = [0]*len(labels)

count = 0
for file in all_file_paths:
    if 'True' in file:
        df = pd.read_csv(file)
        for i,label in enumerate(labels):
            c_rating_in_segment[i] += df[[label in l for l in df['label']]]['c_rating'].mean()
        count += 1
            
print(count)
c_rating_in_segment = np.array(c_rating_in_segment)/count
plt.figure(figsize=(5,3),dpi=200) 
plt.bar(labels, c_rating_in_segment)
# plt.xlabel("")
plt.ylabel("C-Rating (C)")
plt.xticks(rotation=90)
```
